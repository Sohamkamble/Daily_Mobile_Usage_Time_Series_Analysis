```python
app_categories = {
    "Social Media": ["Instagram", "WhatsApp", "Facebook"],
    "Web Browsers": ["Chrome", "Firefox"],
    "Video Streaming": ["YouTube"],
    "Professional Networking": ["LinkedIn"],
    "Communication": ["Call", "Messages", "Gmail"],
    "Gaming": ["Clash Royale", "Dream Cricket"],
    "Productivity": ["Drive", "Calendar", "Sheets", "Files"],
    "Entertainment": ["Spotify", "Gallery", "Camera", "Photos", "Swayam", "Calculator", "Google Play Store", "Internshala", "BookMyShow", "Software update", "Finder"],
    "Utility": ["pod.ai", "Phone", "Amazon", "Maps", "My Files", "Call settings", "Doc Scanner"]
}

# Function to categorize apps
def categorize_apps(app):
    for category, apps in app_categories.items():
        if app in apps:
            return category
    return

# Apply categorization
df1["Category"] = df1["App name"].apply(categorize_apps)

# Display categorized DataFrame
print(df1)
```

```python
category_counts = df1["Category"].value_counts()

# Create bar plot
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('App Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
